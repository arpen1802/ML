## Implement Newton's method with tf and visualize the process

Check [Wikipedia](https://en.wikipedia.org/wiki/Newton%27s_method) and [chapter 8.6 Approximate Second-Order Methods](http://www.deeplearningbook.org/contents/optimization.html).

Find roots with **Newton's method** for:


1) $\:f(x) = x\,cos(x) - x^2 \:\:\:  x_0 = 1$

2) $\:f(x) = x^3 - 7x^2 + 8x - 3 \:\:\: x_0 = 5 $

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def first_function(x):
    return (x*np.cos(x)) - x*x
```

```python
def second_function(x):
    return (pow(x,3)) - (7*pow(x,2)) + (8*x) - 3
```

```python
def newton(f,x0):
    xn = x0
    xn_1 = 0
    points = []
    while(True):
        with tf.GradientTape() as g:
            g.watch(xn)
            y = f(xn)
        dy_dx = g.gradient(y, xn)
        xn_1 = xn - f(xn)/dy_dx
        points.append(xn_1)
        if(abs(xn_1-xn)<0.00001 or tf.math.is_nan(xn_1)):
            tf.print('Root is:',xn_1)
            break
            
        else:
            xn=xn_1
    plt.plot(points)

    
    
```

```python
x = tf.constant(1,dtype=tf.float32)
newton(first_function,x)
```

```python
x = tf.constant(5,dtype=tf.float32)
newton(second_function,x)
```
