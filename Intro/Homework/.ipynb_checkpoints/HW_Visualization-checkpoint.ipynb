````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
evn = pd.read_csv("../Data/yerevan_houses.csv")
evn.head()
````

#### Get rid of Unnamed:0 column

````python
evn = evn.drop(columns=['Unnamed: 0'])
````

#### Catch the NaN values and try to handle them. Possible treatments could be:

* dropping the records containing NaN values
* filling the NaN values with something meaningfull

````python
evn.dropna(inplace=True)
````

#### Find out existing correlations with "price" column. Plot the heatmap.

````python
evn.corr()['price']
````

````python
sns.heatmap(evn.corr())
````

#### Try doing some bivariate analysis for "price" column. <br> For numerical columns you can use jointplot and pairplot, for numerical columns try to make catplots (barplots, countplots, boxplots including).  Use categorical features as hue parameter in pairplots. Include some notes and conclusions you make upon all of these accordingly. For example, do you notice a common pattern between "price" and "district"? Or how does the number of rooms affect on the house price, etc. 

````python
evn_analysis = evn.drop(['url','region','street'],axis=1)
#url and street have mulitple values, and thus are not suitable to be analysed in this task. There is only one region.
check = (evn_analysis.dtypes!='object') & (evn_analysis.columns != 'price')
columns_numerical =  evn_analysis.columns[check]
columns_categorical = evn_analysis.columns[evn_analysis.dtypes=='object']
columns_numerical_with_price =  evn_analysis.columns[evn_analysis.dtypes!='object']
for ax, col in enumerate(columns_numerical):
    sns.jointplot(x=col,y='price',data=evn_analysis,kind='scatter');
````

# 1) As we see different factors have different effects on price change. For example in average houses with 3-4 rooms are more valuable or that number of bathrooms does not play huge role on price change. Also according to data, the most valuable houses are ones with 3-3.2 ceiling height.

````python
for i in columns_categorical:
    sns.pairplot(evn_analysis,hue=i,palette='coolwarm')
````

````python
fig, axes = plt.subplots(3, 1, figsize=(24, 10))
sns.boxplot(x='district', y="price", data=evn_analysis,palette='rainbow',ax=axes[0])
sns.boxplot(x='condition', y="price", data=evn_analysis,palette='rainbow',ax = axes[1])
sns.boxplot(x='building_type', y="price", data=evn_analysis,palette='rainbow')
````

## 2) The district distribution, for native citizens is not so surprising. In average houses in center and in Norq Marash have higher price. The least valueable are houses from Shengavit and Nubarashen. Also, surprisingly, monolit construction have higher price then stone houses.

````python
fig, axes = plt.subplots(3, 1, figsize=(20, 10))
sns.countplot(x='district', data=evn_analysis,palette='rainbow',ax=axes[0])
sns.countplot(x='condition', data=evn_analysis,palette='rainbow',ax = axes[1])
sns.countplot(x='building_type' , data=evn_analysis,palette='rainbow')
````

## 3) In this list of data most of the houses are in Center and Arabkir. Most of them are newly repaired and about 2500 of them are built from stone.

````python
fig, axes = plt.subplots(3, 1, figsize=(20, 10))
sns.barplot(x='district',y='price' ,data=evn_analysis,palette='rainbow',ax=axes[0])
sns.barplot(x='condition',y='price', data=evn_analysis,palette='rainbow',ax = axes[1])
sns.barplot(x='building_type' ,y='price', data=evn_analysis,palette='rainbow')
````

#### Write code which plots distplots of all numeric columns in a loop.


````python
fig, axes = plt.subplots(7, 1, figsize=(6, 15))
for i,col in enumerate(columns_numerical_with_price):
    sns.distplot(evn_analysis[col],kde=True,ax=axes[i]);
````

#### Plot boxplot and barplot couples as subplots for all categorical columns vs "price" in a loop. <br> Use tight layout to avoid overlapping. Use ```fig, axes = plt.subplots(nrows, ncols)``` 

````python
fig, axes = plt.subplots(3, 2, figsize=(24, 10))
for i, col in enumerate(columns_categorical):
    sns.boxplot(x=col, y="price", data=evn_analysis,palette='rainbow',ax=axes[i][0])
    sns.barplot(x=col,y='price' ,data=evn_analysis,palette='rainbow',ax=axes[i][1])
fig.tight_layout()

````

#### **Bonus: Try to visualize the street occurencies in this dataframe on map. Just checking googling skills :)

````python
from geopy.geocoders import Nominatim
locator = Nominatim(user_agent= 'geopy/1.21.0')
````

````python
from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
new = pd.DataFrame()
test = evn.head(20)
new['location'] = (test['street']+', Yerevan, Armenia').apply(geocode)
new['point'] = new['location'].apply(lambda loc: tuple(loc.point) if loc else None)
new[['latitude', 'longitude', 'altitude']] = pd.DataFrame(new['point'].tolist(), index=new.index)
````

## Taking only 20 points

````python
new.dropna(inplace = True)
````

````python
from IPython.display import display
````

````python
import folium
map1 = folium.Map(
    location=[40.177200,44.503490],
    tiles='Yerevan',
    zoom_start=12,
    attr="<a href=https://endless-sky.github.io/>Endless Sky</a>"
)

new.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]]).add_to(map1), axis=1)
map1
````
