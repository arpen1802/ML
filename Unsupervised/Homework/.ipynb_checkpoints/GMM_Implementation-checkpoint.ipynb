# Imports

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import multivariate_normal
from sklearn.cluster import KMeans
import random
from numpy import linalg as LA
```

# Implemenet GMM

```python
class GMM:
    def __init__(self, k,n_iter = 10000,k_means=True):
        """
        Init method
        param k: number of clusters
        """
        self.n_iter = n_iter
        self.k = k
        self.means = []
        self.covs = []
        self.pis = []
        self.gammas = []
        self.path = []
        self.k_means = k_means
    
    
    def e_step(self,X):
        """
        Implement E-step(calculate gammas)
        use from scipy.stats import multivariate_normal
        """
        gammas = np.zeros((X.shape[0],self.k))
        for k in range(self.k):
            gammas[:,k] = self.pis[k] * multivariate_normal.pdf(X, self.means[k,:], self.covs[k],allow_singular=True)
        gammas = gammas / np.sum(gammas, axis=1).reshape(-1, 1)
        self.gammas = gammas
        
        
        
    
    def m_step(self,X):
        """
        M-step (update means,pis,covs)
        """
        covs = np.zeros((self.k, X.shape[1], X.shape[1]))
        nk = np.sum(self.gammas,axis=0).reshape(-1, 1)
        pis = np.mean(self.gammas, axis = 0)
        means = (self.gammas.T @ X) /nk
        for k in range(self.k):
            covs[k,:, :] = ((self.gammas[:,k] * (X - means[k,:]).T) @ (X - means[k,:]))/(np.sum(self.gammas, axis = 0)[k])
        self.covs = covs
        self.pis = pis
    
    
        
    def fit(self,X):
        """
        Main fit method.Need to
        1. Initialize means, pis, covs
        2. E-step (calculate gammas)
        3. M-step (update means,pis,covs)
        4. Repeat to converge(means dosent not change, or loss dosent not)
        """
        if(self.k_means):
            k_means = KMeans(n_clusters=self.k)
            k_means.fit(X)
            self.means = k_means.cluster_centers_
        else:
            self.means = np.array(random.choices(X,k=self.k))
        self.pis = np.zeros((self.k,1))+1/self.k
        identity = np.identity(X.shape[1])
        shape = self.k, X.shape[1], X.shape[1]
        self.covs = np.full(shape, np.cov(X, rowvar = False))
        n=1
        inital_log = 0
        
        for i in range(self.n_iter):
            self.e_step(X)
            self.m_step(X)
            temp = self.log_likelihood(X)
            self.path.append(temp)
            if abs(temp - inital_log) <= 1e-8:
                break
            inital_log = temp
            

    def predict(self,X):
        """
        Calculate probabilites(gammas) for new X 
        param X(nd array): input dataset
        """
        y_probs = np.zeros((X.shape[0], self.k))
        for k in range(self.k):
            y_probs [:,k] = self.pis[k] * multivariate_normal.pdf(X, self.means[k,:], self.covs[k],allow_singular=True)
        y_probs = y_probs/y_probs.sum(axis=1).reshape(-1,1)
        
        
        
        return y_probs
    
    
    def log_likelihood(self,X):
        """
        Calculate and return log-likelihood
        param X(nd array): input dataset
        """
        gammas = 0
        for k in range(self.k):
            gammas+=np.log((self.pis[k]*multivariate_normal.pdf(X,self.means[k,:],self.covs[k],allow_singular=True)))
        return np.sum(gammas)
        
```

## Testing

```python
from scipy.stats import multivariate_normal
import numpy as np 
```

```python
mean1 = np.array([0,0])
cov1 = np.array([[1,-0.8],[-0.8,1]])

mean2 = np.array([1,2.5])
cov2 = np.array([[1,-0.8],[-0.8,1]])

mean3 = np.array([1,5])
cov3 = np.array([[1,-0.8],[-0.8,1]])

n = 300 
np.random.seed(42)

x1, y1 = np.random.multivariate_normal(mean1, cov1, n).T
x2, y2 = np.random.multivariate_normal(mean2, cov2, n).T
x3, y3 = np.random.multivariate_normal(mean3, cov3, n).T

x = np.concatenate((x1,x2,x3)).reshape(-1,1)
y = np.concatenate((y1,y2,y3)).reshape(-1,1)
X = np.concatenate((x,y),axis=1)
```

```python
model = GMM(3,n_iter=10000,k_means=True)
model2 = GMM(3,n_iter=10000,k_means=False)
```

```python
model.fit(X)
model2.fit(X)
```

```python
model.means
```

```python
fig = plt.figure(figsize=(10,10))
x1 = np.arange(0, len(model.path)) 
x2 = np.arange(0, len(model2.path)) 
plt.plot(x1, model.path,'r')
plt.plot(x2, model2.path)
plt.legend(['With K_means','Without K_means'])
```

```python
w, v = LA.eig(model.covs)
```

```python
fig = plt.figure(figsize=(12,12))
plt.scatter(X[:,0],X[:,1],c=np.argmax(model.predict(X),axis=1));
for i in range(v.shape[0]):
    origin = model.means[i]
    vector = v[i]
    plt.quiver(*origin,vector[:,0],vector[:,1], color=['r','b'], scale=21)
plt.show()
```
