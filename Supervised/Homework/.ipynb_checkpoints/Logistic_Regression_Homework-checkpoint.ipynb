### Assignment 1
Implement LogisticRegression class.

```python
class LogisticRegression:
    """
    Logistic Regression Class

    """

    def __init__(self,betta=None):
        """
        Init method
        """
        self.betta = betta
        

        
    def sigmoid(self, X):
        """
        :param X: m x 1+n matrix, where m is the number of samples, n is the number of features
        :return: sigmoid (logistic) of (X,betta)
        """
        sig =1/(1+np.exp(-X@np.transpose(self.betta))) 
        sig = np.minimum(sig, sig.max())  
        sig = np.maximum(sig, sig.min())
        return sig
    def gradient(self, X, Y):
        """
        :param X: m x 1+n matrix, where m is the number of samples, n is the number of features
        :param Y: 1D-array of labels (for example: [1, 0, 0, 1, .., 0])
        :return: gradient of loss function
        """
        
        return ((self.sigmoid(X)-Y).T@X)/len(Y)


    def fit(self, X, Y, num_iter,alpha=0.7):
        """
        :param X: m x 1+n matrix, where m is the number of samples, n is the number of features
        :param Y: 1D-array of labels (for example: [1, 0, 0, 1, .., 0])
        :param alpha: (float) learning rate
        :param num_iter: number of iterations
        :return: optimal betta
        """
        beta_n = np.zeros((1,X.shape[1]))
        beta_n_1  = np.zeros((1,X.shape[1]))
        
        for i in range(num_iter):
            beta_n_1 = beta_n - alpha*self.gradient(X,Y)
            beta_n = beta_n_1
            self.betta = beta_n_1
        return self.betta

    def predict(self, X):
        """
        :param X: m x 1+n matrix, where m is the number of samples, n is the number of features
        :return: 1D-array of predicted labels (for example: [1, 0, 0, 1, .., 0])
        """
        y = np.zeros((X.shape[0],1))
        y[self.sigmoid(X)>0.5] += 1
        return y

    def loss(self, X, Y):
        """
        Method for calculation current loss
        :param X: (np.array,n*p) features
        :param Y: (np.array,n*1) labels
        :return: Loss
        """
        
        loss = (-np.transpose(Y)@np.log(self.sigmoid(X))-np.transpose(1-Y)@np.log(1-self.sigmoid(X))).sum()/len(Y)
        return loss
        
```

### Assignment 2
* Generate and plot classified dataset with decision boundary being the line $ -x_1 + x_2 = 0 $ (this means true betta is [0, -1, 1]). 
* Use logistic regression to classify your dataset and find optimal betta. 
* Plot the found boundary. 
* Print the loss. <br>
##### You should obtain a figure like the one below.

```python
import pandas as pd
```

```python
x1  = np.random.uniform(0, 10, 500)
x2  = np.random.uniform(0, 10, 500)
data = pd.DataFrame([x1,x2]).T
data.columns = ["x1","x2"]
data['label'] = 0
```

```python
data.loc[data['x1']>data['x2'],'label'] = 1
```

## Now lets add some outliers.

```python
random_index = np.random.choice(data.index,size = 30,replace=False)
```

```python
data.iloc[random_index,2] = data.iloc[random_index,2].apply(np.logical_not).astype('int')
```

```python
fig = plt.figure()
plt.scatter(data['x1'],data['x2'],c = data['label'],label = 'x1 vs x2' )
plt.plot(x1, x1 + 0, 'r',label='Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(bbox_to_anchor =(1,1))
fig.tight_layout()
```

```python
beta = np.array((0,-1,1)).reshape(1,-1)
```

```python
model1 = LogisticRegression(beta)
```

```python
x = data[['x1','x2']]
```

```python
x =  np.insert(x.to_numpy(),0,1,axis=1)
```

```python
y = data['label'].to_numpy().reshape(-1,1)
```

```python
b_hat = model1.fit(x,y,1000,0.1)
```

```python
y_hat = (-1/b_hat[0][2])*(b_hat[0][1]*x[:,1]+b_hat[0][0])
```

## If  0 =b0+b1*x1+b2*x2 then x2 = -b1*x1/b2 - b0/b2

```python
fig = plt.figure()
plt.scatter(data['x1'],data['x2'],c = data['label'])
plt.plot(x1, x1 + 0, 'r',label='Decision Boundary')
plt.plot(x[:,1], y_hat, 'g',label='Found Boundary') 
plt.xlabel('x1')
plt.ylabel('x2')
plt.ylim(0, 11)
plt.legend(bbox_to_anchor =(1,1))

fig.tight_layout()
```

```python
loss = model1.loss(x,y)
```
